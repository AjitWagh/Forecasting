```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
from statsmodels.tsa.ar_model import AutoReg
pd.set_option('display.max_colwidth', 40)
pd.set_option('display.max_rows', 400)
```

```python
df = pd.read_csv('C:\\Users\\Ajit Wagh\\Desktop\\TimeSeries\\Brazilian Eccomerce Dataset\\product_categories\\pet_shop.csv')
df.reset_index()
df = df[['order_purchase_date','price']]
```

```python
df = df.set_index('order_purchase_date')
```

```python
X=df.values
print('Shape of data \t',df.shape)
print('Original Dataset:\n', df.head())
print('After extracting only temperature:\n', X)
```

```python
df.plot()
```

```python
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(df['price'], autolag = 'AIC')

print("1. ADF : ",dftest[0])
print("2. P-Value : ",dftest[1])
print("3. Num of Lags : ",dftest[2])
print("4. Num of observations used for ADF Regression and Critical Values Calculation : ",dftest[3])
for key, val in dftest[4].items():
    print("\t",key, ":", val)
```

```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
pacf = plot_pacf(df['price'],lags=25)
acf = plot_acf(df['price'],lags=25)
```

```python
train=X[:len(X)-7]
test=X[len(X)-7:]
```

```python
model=AutoReg(train,lags=10).fit()
```

```python
print(model.summary())
```

```python
print(len(train))
```

```python
pred=model.predict(start=len(train), end=len(X)-1, dynamic=False)
```

```python
from matplotlib import pyplot
pyplot.plot(pred)
pyplot.plot(test, color='red')
print(pred)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
rmse=sqrt(mean_squared_error(test,pred))
```
