```python
df1 = df[['order_purchase_date','price']]
df1 = df1.rename(columns = {'order_purchase_date':'ds','price':'y'})
forecast_df = df1
```

```python
forecast_df.head()
```

```python
forecast_df['ds'] = forecast_df['ds'].str[:7]
```

```python
forecast_df = forecast_df.groupby('ds')['y'].mean().reset_index()
```

```python
forecast_df.shape
```

```python
# Behaviour of data
ax = forecast_df.set_index('ds').plot(figsize=(12, 8))
ax.set_ylabel('Variation of Data')
ax.set_xlabel('Date')
plt.show()
```

```python
forecast_df.dtypes
```

```python
forecast_df['ds']=pd.to_datetime(forecast_df['ds'], errors='coerce')
forecast_df.dtypes
```

```python
forecast_df.head()
```

```python
m = Prophet(yearly_seasonality=True).fit(forecast_df)
#m = Prophet(seasonality_mode='multiplicative', mcmc_samples=300).fit(forecast_df)
#m = Prophet(seasonality_mode='multiplicative').fit(forecast_df)
#m = Prophet().fit(forecast_df)
future = m.make_future_dataframe(periods=12, freq='MS')
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
data_frames = [forecast_df,fcst]
forecast_data_ndays = reduce(lambda left,right: pd.merge(left,right, on=['ds'],how='outer'),data_frames)
```

```python
forecast_data_ndays.shape
```

```python
forecast_data_ndays['product_category'] = 'fashion_esporte'
forecast_data_ndays= forecast_data_ndays.rename(columns = {'ds':'order_purchase_date','y':'price'})
final_forecast_df = forecast_data_ndays[['order_purchase_date','product_category','price','yhat']]
```

```python
## MAPE
final_forecast = final_forecast_df.iloc[:-12,:]
a = np.mean(np.abs(((final_forecast['price']) - (final_forecast['yhat'])) / (final_forecast['price']))) * 100
```

```python
a
```

```python
final_forecast_df['MAPE']=a
```

```python
final_forecast_df.to_csv('C:\\Users\\Ajit Wagh\\Desktop\\TimeSeries\\new_categories\\new_files\\results\\extra\\fashion_esporte_results.csv',index=False)
```
